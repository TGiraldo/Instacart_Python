```python
#Define a function

def price_label(row):

  if row['prices'] <= 5:
    return 'Low-range product'
  elif (row['prices'] > 5) and (row['prices'] <= 15):
    return 'Mid-range product'
  elif row['prices'] > 15:
    return 'High range'
  else: return 'Not enough data'
```

```python
#Apply the function
ords_prods_merge_1mil['price_range'] = ords_prods_merge_1mil.apply(price_label,axis=1)
```

```python
#Counting the values in new column
ords_prods_merge_1mil['price_range'].value_counts(dropna = False)
```

```python
# New labels to show second most busy and slow days 
result = []

for value in ords_prods_merge["orders_day_of_week"]:
    if value == 0:
        result.append("Busiest day")
    elif value == 1:
        result.append("Busiest day")
    elif value == 4:
        result.append("Slowest day")
    elif value == 3:
        result.append("Slowest day")
    else:
        result.append("Regularly busy")

ords_prods_merge["busiest_days"] = result
```

```python
# Create a list to store the labels for each hour
periods = []

# Create thresholds for most, average, and fewest orders
most_orders_hours = [10, 11, 14, 15,14]
fewest_orders_hours = [4, 3, 2, 5, 1, 0]

# Label each 'order_hour_of_day' column
for hour in ords_prods_merge['order_hour_of_day']:
    if hour in most_orders_hours:
        periods.append("Most_orders")
    elif hour in fewest_orders_hours:
        periods.append("Fewest_orders")
    else:
        periods.append("Average_orders")

# Create a new column in df with the labels
ords_prods_merge['busiest_period_of_day'] = periods
```
